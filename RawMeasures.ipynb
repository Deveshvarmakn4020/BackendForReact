***Binarizer***

```python
from sklearn.preprocessing import Binarizer
dib_df = pd.read_csv('/content/drive/MyDrive/MLDataset/diabetes.csv')
bn = Binarizer(threshold=120)

glucose_binarized = bn.transform([dib_df['Glucose']])[0]
dib_df['Glucose Binarized'] = glucose_binarized

print(dib_df[['Pregnancies','Glucose','Glucose Binarized']])
```

***Rounding***

```python
from sklearn.preprocessing import Binarizer
iris_df = pd.read_csv('/content/drive/MyDrive/MLDataset/Iris.csv')
iris_df['Petal Length Rounded'] = np.array(np.round(iris_df['PetalLengthCm']))
print(iris_df[['Id','SepalLengthCm','PetalLengthCm','Petal Length Rounded']])
```

***Interactions***

```python
from sklearn.preprocessing import PolynomialFeatures
dib_df = pd.read_csv('/content/drive/MyDrive/MLDataset/diabetes.csv')
glucose_pressure = dib_df[['Glucose','BloodPressure']]
pf = PolynomialFeatures(degree=2,interaction_only= False, include_bias = False)
res = pf.fit_transform(glucose_pressure)
res

int_features = pd.DataFrame(res,columns=['Glucose','BloodPressure','Glucose^2','Glucose x BloodPressure', 'BloodPressure^2'])
int_features.head(5)
```

***Fixed Width Binning***

```python
# Sample FIFA 23 dataset
fifa_df = pd.read_csv("/content/drive/MyDrive/MLDataset/fifa23.csv").iloc[:50]
age_bins = np.arange(15,45,5)
bin_names = [1,2,3,4,5]
fifa_df['AgeBin'] = pd.cut(fifa_df['Age'], bins=age_bins, right=False)
fifa_df['AgeBinLable'] = pd.cut(fifa_df['Age'], bins=age_bins, right=False, labels=bin_names)
fifa_df[['Known As','Age','AgeBin','AgeBinLable', 'Nationality']]
```

***Quantile Binning***

```python
fifa_df['AgeQuantileBin'] = pd.qcut(fifa_df['Age'], q=4, labels = ["Q1","Q2","Q3","Q4"])
fifa_df['AgeQuantileBinRange'] = pd.qcut(fifa_df['Age'], q=4)
fifa_df[['Known As','Age','AgeBin','AgeQuantileBin','AgeQuantileBinRange','Nationality']]
```

***Long Transform***

```python
fifa_df['Log Value'] = np.log(fifa_df['Value(in Euro)'])
fifa_df[['Known As','Nationality','Value(in Euro)', 'Log Value']]
```

***Box-Cox Transform***

```python
from scipy import stats
fifa_df['BoxCoxTransformed'], fitted_lambda = stats.boxcox(fifa_df['Value(in Euro)'])
print("Fitted Lambda:", fitted_lambda)
fifa_df[['Known As','Age','Nationality','Value(in Euro)','Log Value','BoxCoxTransformed']]
```
